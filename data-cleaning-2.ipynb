```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_path="train.csv"
df=pd.read_csv(dataset_path)
```

```python
df.isnull().sum()
```

```python
missing_value_percentage=df.isnull().sum() / df.shape[0]*100
missing_value_percentage
```

```python
missing_value_percentage_greater_20=missing_value_percentage[missing_value_percentage>20].keys()
missing_value_percentage_greater_20
```

```python
df.shape
```

```python
df_drop_col=df.drop(columns=missing_value_percentage_greater_20)
df_drop_col.shape
```

```python
df_num=df_drop_col.select_dtypes(include=['int64', 'float64'])
df_num.head()
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(df_num.isnull())
```

```python
df_num[df_num.isnull().any(axis=1)]
```

```python
df_num.isnull().sum()
```

```python
missing_num_var=[var for var in df_num.columns if df_num[var].isnull().sum()>0]
missing_num_var
```

```python
df_num.isnull().sum() / df_num.shape[0]*100
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.displot(df_num[var], bins=20, kde_kws={'linewidth':5, 'color': '#DC143C'})
```

```python
df_num.isnull().sum().sum()
```

```python
df_num_mean=df_num.fillna(df_num.mean())
df_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.displot(df_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color': 'green'})
```

```python
df_num_median=df_num.fillna(df_num.median())
df_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.displot(df_num_median[var], bins=20, kde_kws={'linewidth':5, 'color': 'green'})
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.displot(df[var], bins=20, kde_kws={'linewidth':5, 'color': 'green'})
    sns.displot(df_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color': 'blue'}, label='Mean')
    sns.displot(df_num_median[var], bins=20, kde_kws={'linewidth':5, 'color': 'red'}, label='Median')
```

```python

for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10, 10))
    plt.subplot(3, 1, 1)
    sns.boxplot(df[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(df_num_mean[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(df_num_median[var])
```
