```python
df_copy=df.copy()
for var_class in df['LotConfig'].unique():
      df_copy.update(df[df.loc[:, 'LotConfig']==var_class]['LotFrontage'].replace(np.nan, df[df.loc[:, 'LotConfig']=='Inside']['LotFrontage'].mean()))
```

```python
df_copy.isnull().sum()
```

```python
df_copy=df.copy()
num_vars=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig', 'MasVnrType', 'GarageType']
for num_var, cat_var in zip(num_vars, cat_vars):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var]==var_class][num_var].replace(np.nan, df[df.loc[:, cat_var]==var_class][num_var].mean()))
```

```python
df_copy[num_vars].isnull().sum()
```

```python
df_copy[df_copy[['MasVnrArea']].isnull().any(axis=1)]
```

```python
df_copy[df_copy[['GarageType']].isnull().any(axis=1)]
```

```python
df_copy=df.copy()
num_vars=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig', 'Exterior2nd', 'KitchenQual']
for num_var, cat_var in zip(num_vars, cat_vars):
    for var_class in df[cat_var].unique():
        df_copy.update(df[df.loc[:, cat_var]==var_class][num_var].replace(np.nan, df[df.loc[:, cat_var]==var_class][num_var].mean()))
```

```python
df_copy[num_vars].isnull().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(num_vars):
    plt.subplot(2,2,i+1)
    sns.displot(df[var], bins=20, kde_kws={'linewidth':8, 'color': 'blue'})
    sns.displot(df_copy[var], bins=20, kde_kws={'linewidth':5, 'color': 'red'})
    plt.legend()
```

```python
df_copy_median=df.copy()
num_vars=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars=['LotConfig', 'Exterior2nd', 'KitchenQual']
for num_var, cat_var in zip(num_vars, cat_vars):
    for var_class in df[cat_var].unique():
        df_copy_median.update(df[df.loc[:, cat_var]==var_class][num_var].replace(np.nan, df[df.loc[:, cat_var]==var_class][num_var].median()))
```

```python
df_copy_median.isnull().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(num_vars):
    plt.subplot(2,2,i+1)
    sns.displot(df[var], bins=20, kde_kws={'linewidth':8, 'color': 'blue'})
    sns.displot(df_copy[var], bins=20, kde_kws={'linewidth':5, 'color': 'red'})
    sns.displot(df_copy_median[var], bins=20, kde_kws={'linewidth':5, 'color': 'green'})
    plt.legend()
```
