```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
cat_vars=df.select_dtypes(include='object')
cat_vars
```

```python
miss_val_per=cat_vars.isnull().mean()*100
miss_val_per
```

```python
cat_vars_miss=cat_vars.isnull().sum()/cat_vars.shape[0]*100
cat_vars_miss
```

```python
cat_vars.isnull().sum()
```

```python
missing_val_gre_20=cat_vars_miss[cat_vars_miss>20].keys()
missing_val_gre_20
```

```python
cat_vars.shape
```

```python
drop_cat_column = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
cat_vars.drop(columns=drop_cat_column, axis=1, inplace=True)
cat_vars.shape
```

```python
isnull_per=cat_vars.isnull().mean()
miss_vars=isnull_per[isnull_per>0].keys()
miss_vars
```

```python
cat_vars['MasVnrType'].fillna('Missing')
```

```python
cat_vars['MasVnrType'].mode()
```

```python
cat_vars['MasVnrType'].value_counts()
```

```python
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0])
```

```python
for var in miss_vars:
    cat_vars[var].fillna(cat_vars[var].mode()[0], inplace=True)
    print(var, '=', cat_vars[var].mode()[0])
```

```python
cat_vars.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i, var in enumerate(miss_vars):
    plt.subplot(4,3,i+1)
    plt.hist(cat_vars[var], label='Imput')
    plt.hist(df[var].dropna(), label='Original')
    plt.legend()
```
