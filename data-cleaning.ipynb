```python
pd.concat([df['MSZoning'].value_counts() / df.shape[0] * 100,
          df2_drop_column['MSZoning'].value_counts() / df2_drop_column.shape[0] * 100], axis=1, keys=['MSZoning_org', 'MSZoning_clean'])
```

```python
def cat_var_dist(var):
    return pd.concat([df[var].value_counts() / df.shape[0] * 100,
    df2_drop_column[var].value_counts() / df2_drop_column.shape[0] * 100], axis=1, keys=[var+'_org', var+'_clean'])
```

```python
cat_var_dist('MSZoning')
```
